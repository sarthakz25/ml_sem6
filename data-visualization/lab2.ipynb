```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = np.random.uniform(0, 10, 5)  # 5 random numbers from 0 to 10
Y = np.random.uniform(11, 21, 5)  # 5 random numbers from 11 to 21
```

```python
plt.scatter(X, Y)
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('scatter plot of random values')
plt.savefig('scatter_plot.png')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(X, Y, c='blue', marker='o', label='blue circles')
ax.scatter(X, Y+10, c='red', marker='^', label='red triangles')
ax.legend()
plt.show()
```

```python
plt.scatter(df['Kernel.Length'], df['Kernel.Width'], c="silver")
```

```python
import seaborn as sns
sns.scatterplot(x='Kernel.Length', y='Kernel.Width', data=df, c='g')
```

```python
sns.scatterplot(x='Kernel.Length', y='Kernel.Width', hue='Type', data=df)

#  inference: type 3 has lowest length and width
#             type 2 has highest length and width
#             there is an overlap between type 3 & 1, and type 1 & 2
```

```python
sns.jointplot(x='Perimeter', y='Compactness', data=df)

# inference -
# no major outliers are present
# joint distributions have similar spreads, relationship is similar
```

```python
sns.boxplot(x="Type", y="Compactness", data=df)
```

```python
sns.pairplot(data=df, hue="Type", diag_kind='hist')

# inference -
# perimeter correlated strongly with length and width
# Some characteristics more distinct for certain types
```

```python
sns.violinplot(y='Compactness', x='Type', data=df)

# inference -
# type 1 is concentrated between a certain range
# type 3 compactness starts at less than 0.83
# median of type 1 and type 2 is quite similar i.e. overlaping
```

```python
sns.FacetGrid(df, hue='Type', height=7).map(sns.kdeplot, 'Compactness').add_legend()

# inference -
# type 1 most compact, type 3 least compact
# distributions differ visibly across types
```

```python
# kernel density estimation plot
sns.pairplot(data=df, hue="Type")
```
