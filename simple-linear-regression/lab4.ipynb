### **Task 1**

```python
# 1.	Define two Numpy arrays x and y   which represents BMI and Cholesterol
# x= [5,15,25,35,45,55] and y= [11,16,18,30,22,38].
# 2.	Plot a scatter plot of x and y.
# 3.	Write a python function to compute the values of the coefficients of linear regression, b0 and b1.
# 4.	Determine the predicted value of y for x=27.
# 5.	Plot the regression line on the scatter plot.
# 6.	Write a python function to determine the value of r2.
```

```python
import numpy as np
import pandas as pd
```

```python
bmi= [5,15,25,35,45,55]
chol= [11,16,18,30,22,38]

x = np.array(bmi)
y = np.array(chol)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x, y, c="silver")
plt.xlabel('BMI')
plt.ylabel('Cholesterol')
```

```python
def linear_regression(x,y):
    n = len(x)
    sum_x = sum(x)
    sum_y = sum(y)
    sum_xy = sum(x*y)
    sum_xx = sum(x*x)

    b1 = (n*sum_xy - sum_x*sum_y) / (n*sum_xx - sum_x**2)
    b0 = (sum_y - b1*sum_x)/n

    return b0, b1
```

```python
b0, b1 = linear_regression(x, y)
print(b0, b1)
```

```python
y_pred = b0 + b1*27
print(y_pred)
```

```python
plt.scatter(x,y)
plt.plot(x, b0 + b1*x, color='y')
```

```python
def r_squared(y, y_pred):
    y_mean = np.mean(y)
    ss_tot = sum((y - y_mean)**2)
    ss_res = sum((y - y_pred)**2)
    r2 = 1 - (ss_res/ss_tot)
    return r2

r2 = r_squared(y, b0 + b1*x)
print(r2)
```
